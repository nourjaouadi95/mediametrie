# 1. Exploser le path

```python
stop_words = set(stopwords.words('french'))
```

```python
import math
def to_bag_of_word(path):
    words = ['no_path']
    if path != "" and path == path:
        words = re.findall("([a-zA-Z]+)",path)
        
    words = list(filter(None, words))
    
    retour = []
    
    for word in words:
        if(len(word)>1):
            retour.append(word)
    
    retour = [w for w in retour if not w in stop_words]
    
    return(retour)
```

```python
cookies['path_explode'] = cookies.apply(lambda row : to_bag_of_word(row['path']), axis = 1)
```

# 3. Cration de la table users (Aggrgation par cookie_id)

```python
def col_of_list_to_list(col):
    bag_of_words = col.tolist()
    bag_of_words = [item for sublist in bag_of_words for item in sublist]
    bag_of_words = set(bag_of_words)
    return(bag_of_words)
```

```python
users_and_path = cookies.groupby('COOKIE_ID').apply(lambda subf: col_of_list_to_list(subf['path_explode']))
```

```python
users_and_path = users_and_path.to_frame()
users_and_path['COOKIE_ID'] = users_and_path.index
```

```python
users_and_domains = cookies.groupby('COOKIE_ID').apply(lambda subf: set(subf['netloc'].tolist()))
```

```python
users_and_domains = users_and_domains.to_frame()
users_and_domains['COOKIE_ID'] = users_and_domains.index
```

```python
users_and_path.columns = ['words_in_path','COOKIE_ID']
```

```python
users_and_path.head()
```

```python
users_and_domains.columns = ['domains','COOKIE_ID']
```

```python
users_and_domains.head()
```

```python
users = pd.merge(users_and_path, users_and_domains, on='COOKIE_ID', how='inner')
```

```python
users.head()
```

```python
users.to_csv('users.csv',index=False)
```
